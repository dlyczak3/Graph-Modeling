## Graph Interpretation:

#### Each node represents one of the [size] most connected courses to your specified course

#### The thickness of the line connecting two nodes represents how connected they are:
##### - Thicker means they are very connected (many students take one and then the other)
##### - Thinner means few(er) students take the courses in subsequent semesters
#### The color of the node, if lighter, shows that that course is not as connected to the other courses
